# Resumo dos arquivos CSV – pasta dados

Este notebook analisa os arquivos CSV da pasta `dados/` do projeto (Relatório Fat / vendas).

```python
from pathlib import Path

# Funciona com cwd = raiz do projeto ou pasta notebook/
pasta_dados = Path('dados') if Path('dados').exists() else Path('../dados')
arquivos = sorted(pasta_dados.glob('*.csv'))
print(f'Arquivos CSV encontrados: {len(arquivos)}')
for f in arquivos:
    print(f'  - {f.name}')
```

```python
def resumo_csv(caminho):
    p = Path(caminho)
    if not p.exists():
        return None
    tamanho_mb = p.stat().st_size / (1024 * 1024)
    enc = 'utf-8-sig'
    try:
        with open(p, 'r', encoding=enc) as f:
            linha1 = f.readline()
    except UnicodeDecodeError:
        enc = 'latin-1'
        with open(p, 'r', encoding=enc) as f:
            linha1 = f.readline()
    sep = ';' if ';' in linha1 else ','
    colunas = [c.strip() for c in linha1.split(sep)]
    with open(p, 'r', encoding=enc, errors='ignore') as f:
        n_linhas = sum(1 for _ in f)
    n_registros = max(0, n_linhas - 1)
    return {
        'Arquivo': p.name,
        'Registros': n_registros,
        'Linhas (c/ cabeçalho)': n_linhas,
        'Colunas': len(colunas),
        'Tamanho (MB)': round(tamanho_mb, 2),
        'Separador': sep,
    }

resumos = [resumo_csv(arq) for arq in arquivos if resumo_csv(arq)]
```

```python
import pandas as pd

pd.DataFrame(resumos)
```
